```python
import requests
from io import StringIO
import pandas as pd
import json

# https://api.coinalyze.net/v1/funding-rate-history

url = ("https://api.coinalyze.net/v1/funding-rate-history?symbols=BTCUSDT_PERP.3&interval=12hour&from=1489567282&to=1715805040")

payload={}
headers = {
    'api_key': '44ef6fbb-24fa-4b9a-a640-c01a901a3b60'
}

response = requests.request("GET", url, headers=headers, data=payload)
print(response.text[:1000000])

# Parse JSON string into Python object
data = json.loads(response.text)

# Extract the 'history' attribute
history_data = data[0]['history']

df = pd.DataFrame(history_data)
df.head()
df.info()
```

```python
df['date'] = pd.to_datetime(df['t'],unit='s')
df = df.rename(columns={"o": "open", "h": "high", "l": "low", "c": "close"})
df.head()
```

```python
df = df[["date", "close"]]
df.to_csv("data/BTC_OKX.csv")
```
